```python
# Divida o conjunto de dados em um conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(df[features], df["price"], test_size=0.25)
```

```python


# Selecione as 5 características mais importantes
selector = SelectKBest(score_func=chi2, k=5)
X_train_selected = selector.fit_transform(X_train, y_train)
X_test_selected = selector.transform(X_test)

# Crie um modelo de regressão linear
model = LinearRegression()
```

```python

# Treine o modelo
model.fit(X_train_selected, y_train)

# Preveja os preços das casas no conjunto de teste
y_pred = model.predict(X_test_selected)

```

```python
# Avalie a precisão do modelo
print("Accuracy:", model.score(X_test_selected, y_test))
```
